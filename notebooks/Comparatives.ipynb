```python

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

file_path= './datasets/all_results'
# Set the style of the visualization
sns.set(style="whitegrid")

# Define the metrics to compare
metrics = ['accuracy', 'f1', 'gmean']
var_predicted= ['lent', 'amp']
over_sampling= ['SMOTE', 'ROS']
# Create a boxplot for each metric
for var in var_predicted:
    for over in over_sampling:
        
        data = pd.read_csv(f'{file_path}_{over}_{var}.csv')
        plt.figure(figsize=(15, 10))
        for i, metric in enumerate(metrics, 1):
            plt.subplot(3, 1, i)
            sns.boxplot(x='model', y=metric, data=data)
            plt.title(f'Comparación de modelos mediante {metric.capitalize()} con {over} y {var}')
            plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()
```
